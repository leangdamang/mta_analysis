```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import datetime as dt

df = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170624.txt')
df = df.append(pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170617.txt'), ignore_index = True)
df = df.append(pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170610.txt'), ignore_index = True)
df = df.append(pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170603.txt'), ignore_index = True)
```

```python
df.columns = [columns.strip() for columns in df.columns]
```

```python
df['PERIOD'] = pd.to_datetime(df['DATE'] + ' ' + df['TIME'])
```

```python
df.sort_values(['C/A', 'UNIT', 'SCP', 'STATION', 'PERIOD'])
```

```python
(df.groupby(["C/A", "UNIT", "SCP", "STATION", "PERIOD"]).ENTRIES.count().reset_index().sort_values("ENTRIES", ascending=False)).head(5)
```

```python
turnstile = ((df["C/A"] == "R518") & 
(df["UNIT"] == "R261") & 
(df["SCP"] == "00-03-02") & 
(df["STATION"] == "40 ST LOWERY ST") &
(df["PERIOD"] == '2017-06-05 08:00:00'))

df[turnstile].head(5)
#it looks like it's an adjustment of only 1
```

```python
df.drop_duplicates(subset=["C/A", "UNIT", "SCP", "STATION", "PERIOD"], inplace=True)
```

```python
df_new = df.drop(['DIVISION', 'DATE', 'TIME'], axis = 1)
df_new.groupby(['C/A','UNIT','SCP','STATION', 'PERIOD'])['ENTRIES'].sum() 
```

```python
df_new['DAILYENTRY'] = df_new.groupby(['C/A','UNIT','SCP', 'STATION'])['ENTRIES'].transform(lambda x: x.diff())
```

```python
df_new['DAILYEXIT'] = df_new.groupby(['C/A','UNIT','SCP', 'STATION'])['EXITS'].transform(lambda x: x.diff())
```

```python
df_new[df_new['DAILYENTRY'] < 0].head(10)
```

```python
df_new.describe()
```

```python
df_new.loc[df_new['DAILYENTRY'] < 0 , 'DAILYENTRY'] = df_new['DAILYENTRY'].apply(lambda x: -x)
```

```python
df_new.loc[df_new['DAILYEXIT'] < 0 , 'DAILYEXIT'] = df_new['DAILYEXIT'].apply(lambda x: -x)
```

```python
#outlier_entry = (df_new['DAILYENTRY'].quantile(.75) - df_new['DAILYENTRY'].quantile(.25))*1.5
#outlier_exit = (df_new['DAILYEXIT'].quantile(.75) - df_new['DAILYEXIT'].quantile(.25))*1.5

#df_new.loc[df_new['DAILYENTRY'] > outlier_entry, 'DAILYENTRY'] = df_new['DAILYENTRY'].median()
#df_new.loc[df_new['DAILYEXIT'] > outlier_entry, 'DAILYEXIT'] = df_new['DAILYEXIT'].median()

df_new.describe()
```

```python
df_new.loc[df_new['DAILYENTRY'] > 1000, 'DAILYENTRY'] = df_new['DAILYENTRY'].apply(lambda x: x%1000)
df_new.loc[df_new['DAILYEXIT'] > 1000, 'DAILYEXIT'] = df_new['DAILYEXIT'].apply(lambda x: x%1000)
```

```python
df_station = df_new.groupby(['STATION', 'LINENAME', 'PERIOD'])['DAILYENTRY', 'DAILYEXIT'].sum()
```

```python
df_station=df_station.reset_index()
df_station.head()
```

```python
weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_station['WEEKDAY'] = df_station['PERIOD'].dt.weekday 
df_station = df_station.rename(columns = {'PERIOD': 'HOURS'})
```

```python
df_station[df_station['WEEKDAY'] == 5]
```

```python
df_station.loc[(df_station['HOURS'].dt.hour == 0), 'DAILYENTRY'] = df_station['DAILYENTRY'].median()
```

```python
"""creates heatmaps of stations"""

plt.figure(figsize=(20,10))
df_wtc = df_station.loc[(df_station['STATION'] == 'PATH NEW WTC') & (df_station['LINENAME'] == '1')]
df_wtc = df_wtc.groupby(['WEEKDAY', df_station['HOURS'].dt.hour])['DAILYENTRY'].sum()
df_wtc = df_wtc.reset_index()
df_wtc_piv = df_wtc.pivot('HOURS', 'WEEKDAY','DAILYENTRY')
ax = sns.heatmap(df_wtc_piv, xticklabels = weekday, yticklabels = 2, linewidths=.5, cbar = False)
ax.set_title('WTC', fontsize = 20)
```
