```python
from __future__ import print_function

import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import datetime as dt

#challenge 1 - loading and grouping by first 4 columns

df = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170624.txt')
df = df.append(pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170617.txt'), ignore_index = True)
df = df.append(pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170610.txt'), ignore_index = True)
df = df.append(pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170603.txt'), ignore_index = True)

df.columns = [columns.strip() for columns in df.columns]

df_grouped = df.groupby(['C/A', 'UNIT', 'SCP', 'STATION']) 

df_grouped.head()
```

```python
#challenge 2 - changing date/time to datetime and getting rid of extra columns

df['PERIOD'] = pd.to_datetime(df['DATE'] + ' ' + df['TIME'])
df_new = df.drop(['DIVISION', 'DATE', 'TIME', 'DESC', 'EXITS'], axis = 1)
df_new.groupby(['C/A','UNIT','SCP','STATION', 'PERIOD'])['ENTRIES'].sum() 
```

```python
#challenge 3 - get daily counts for a turnstile
df_new['DAILYENTRY'] = df_new.groupby(['C/A','UNIT','SCP', 'STATION'])['ENTRIES'].transform(lambda x: x.diff())
"""
Getting rid of outliers and negative values
"""
df_new.loc[df_new['DAILYENTRY'] < 0 , 'DAILYENTRY'] = df_new['DAILYENTRY'].apply(lambda x: -x)
df_new.loc[df_new['DAILYENTRY'] > 1000, 'DAILYENTRY'] = df_new['DAILYENTRY'].apply(lambda x: x%1000)

turnstile = ((df_new['C/A'] == 'A002') 
        & (df_new['UNIT'] =='R051') 
        & (df_new['SCP'] == '02-00-00') 
        & (df_new['STATION'] == '59 ST'))
df_unittime = df_new[turnstile].groupby(df_new['PERIOD'].dt.date)['DAILYENTRY'].sum()
```

```python
#challenge 4 - plot turnstile daily 
#it looks like there are a lot of negative values - I would clean it up by either getting rid of them or 
#taking the abs value. I also know there are lots of very high values for daily, so would get rid of these as well 


df_unittime.plot(figsize = (10,10), title = "Daily Entries for 59th St A002 R051 02-00-00 Unit")
```

```python
#CHALLENGE 5 - GROUPING BY TURNSTILES

df_turnstile = df_new.groupby(['STATION', 'LINENAME', 'SCP', df_new['PERIOD'].dt.date])['DAILYENTRY'].sum()
df_turnstile.head()
```

```python
#CHALLENGE 6 - GROUPING BY STATION
df_stationdata = df_new.groupby(['STATION', 'LINENAME', df_new['PERIOD'].dt.date])['DAILYENTRY'].sum()
df_stationdata.head()
df_stationdata = df_stationdata.reset_index()
```

```python
#CHALLENGE 7 - PLOTTING A STATION 

station = ((df_stationdata['STATION'] == 'CHAMBERS ST') & (df_stationdata['LINENAME'] == 'ACE23'))
ax = df_stationdata[station].plot(figsize = (10,10), title = "Daily Entries for Chambers St")
ax.set(xlabel = 'Date', ylabel='Entries')
```

```python
#CHALLENGE 8 - RAINBOW PLOT OF WEEKLY RIDERSHIP
weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_stationdata['PERIOD']= pd.to_datetime(df_stationdata['PERIOD'])
df_stationdata['WEEKDAY'] = df_stationdata['PERIOD'].dt.weekday 
df_stationdata = df_stationdata.sort_values(['STATION', 'LINENAME', 'PERIOD'])
df_chamber = df_stationdata[station]

#want to get weekly slices and plot all on the same graph
plt.plot(df_chamber['WEEKDAY'][2:9], df_chamber['DAILYENTRY'][2:9])
plt.plot(df_chamber['WEEKDAY'][9:16], df_chamber['DAILYENTRY'][9:16])
plt.plot(df_chamber['WEEKDAY'][16:23], df_chamber['DAILYENTRY'][16:23])
plt.title('Weekly Activity for Chambers')
```

```python
#CHALLENGE 9 - SUM TOTAL READERSHIP PER STATION AND SORT 

df_tot_station = df_stationdata.groupby(['STATION', 'LINENAME'])['DAILYENTRY'].sum()
df_tot_station = df_tot_station.reset_index()
df_tot_station = df_tot_station.sort_values('DAILYENTRY', ascending = False)
df_tot_station.head(10)
```
